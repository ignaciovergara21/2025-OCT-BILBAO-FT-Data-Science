## Ejercicio Extra
Programa una calculadora que

* Pregunte qu√© operaci√≥n quieres realizar (multiplicaci√≥n/divisi√≥n/suma/resta/potencia/salir)
* Pida dos n√∫meros 
* Realice la operaci√≥n
* Siga funcionando hasta que el usuario elija salir

#### Bonus:
* Que guarde todas las operaciones realizadas y las imprima por pantalla cuando el usuario elija esa opci√≥n
* Que sea robusta frente a entradas err√≥neas del usuario o frente a operaciones como dividir por 0

```python
import math
import pprint
import time
```

```python
# Mostramos el men√∫ inicial con instrucciones
print("Hola, ¬øqu√© operaci√≥n quieres realizar?")
time.sleep(0.1)
print("Opciones disponibles:")
time.sleep(0.1)
print()
time.sleep(0.1)
print("1 - Multiplicaci√≥n ‚úñ")
time.sleep(0.1)
print("2 - Divisi√≥n ‚ûó")
time.sleep(0.1)
print("3 - Suma ‚ûï")
time.sleep(0.1)
print("4 - Resta ‚ûñ")
time.sleep(0.1)
print("5 - Potencia üìà")
time.sleep(0.1)
print()
time.sleep(0.1)
print("Escribe 'salir' üö™ para terminar.")
time.sleep(0.1)
print("Escribe la palabra 'resultados' para poder verlos üëÄ")
time.sleep(0.1)
print()
time.sleep(0.1)


# Inicializamos la variable usuario con una cadena vac√≠a
usuario = ""

# Lista vac√≠a para almacenar los resultados de las operaciones
resultados= []

# Bucle que se ejecuta mientras el usuario no escriba "salir"
while usuario != "salir":
    # Pedimos la operaci√≥n al usuario
    usuario = input("\nElige una opci√≥n (1/2/3/4/5), escribe 'salir' üö™ o escribe 'resultados' si quieres ver el total de todas las operaciones: ")

    # Si el usuario quiere salir, terminamos el bucle
    if usuario.lower() == "salir":
        print("Gracias por usar la calculadora. ¬°Hasta luego! üëã")
        break  # Sale del bucle
    
    # hacemos lo mismo con resultados, cuando el usuarioescribe esa palabra, se para el bucle 
    if usuario.lower() == "resultados":
        print ("Aqu√≠ tienes los resultados de tus operaciones üëÄ:")
        #con el pprint nos formatea el resultado para visualizarlo mejor
        pprint.pprint(resultados)
        break #salimos del bucle para imprimir los resultados
    
    
    # Validamos que la opci√≥n ingresada sea un n√∫mero v√°lido
    if usuario not in ["1", "2", "3", "4", "5"]:
        print("‚ùå Opci√≥n no v√°lida. Escribe un n√∫mero entre 1 y 5 o 'salir'.")
        continue  # Volver a pedir la opci√≥n sin ejecutar m√°s c√≥digo en esta iteraci√≥n

    # Convertimos la opci√≥n a entero
    usuario = int(usuario)
    
    #tengo que volver a validar los n√∫meros
    entrada_1= input("Ingresa el primer n√∫mero:")
    #para ello he probado el m√©todo isdigit
    #Pedimos el primer n√∫mero y validamos que sea un n√∫mero positivo
    if not entrada_1.isdigit():
        print("‚ùå N√∫mero no v√°lido. Escribe un n√∫mero positivo.")
        continue
    num_1 = int(entrada_1)
    
    # Pedimos el segundo n√∫mero y validamos que sea un n√∫mero positivo
    entrada_2= input("Ingresa el segundo n√∫mero:")
    if not entrada_2.isdigit():
        print("‚ùå N√∫mero no v√°lido. Escribe un n√∫mero positivo.")
        continue
    num_2 = int(entrada_2)
 

    # Realizamos la operaci√≥n correspondiente seg√∫n la opci√≥n escogida
    if usuario == 1:
        resultado = num_1 * num_2
        print(f"El resultado de {num_1} ‚úñ {num_2} es {resultado}")
        #con este m√©todo (append) permite que m√°s tarde, cuando el usuario elija ver los resultados, 
        #la lista resultados se imprima con todas las operaciones realizadas
        resultados.append(f"multiplicaci√≥n {num_1} x {num_2} = {resultado}")

    #En la divisi√≥n comprobamos que el usuario no divida entre 0
    elif usuario == 2:
        if num_2 == 0:
            print("‚ùå Error: No se puede dividir entre 0.")
        else:
            resultado = num_1 / num_2
            print(f"El resultado de {num_1} ‚ûó {num_2} es {resultado}")
            resultados.append(f"divisi√≥n {num_1} ‚ûó {num_2} = {resultado}")
            
    elif usuario == 3:
        resultado = num_1 + num_2
        print(f"El resultado de {num_1} ‚ûï {num_2} es {resultado}")
        resultados.append(f"suma {num_1} ‚ûï {num_2} = {resultado}")

    elif usuario == 4:
        resultado = num_1 - num_2
        print(f"El resultado de {num_1} ‚ûñ {num_2} es {resultado}")
        resultados.append(f"suma {num_1} ‚ûñ {num_2} = {resultado}")


    elif usuario == 5:
        resultado = num_1 ** num_2
        print(f"El resultado de {num_1} üìà {num_2} es {resultado}")
        resultados.append(f"suma {num_1} üìà {num_2} = {resultado}")
```
